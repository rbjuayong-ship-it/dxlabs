```python
# get all even numbers between 1 and 50 and print each in word form
def num_to_words(n):
    units = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine", "ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen", "sixteen", "seventeen", "eighteen", "nineteen"]
    tens = ["", "", "twenty", "thirty", "forty", "fifty"]
    if n < 20:
        return units[n]
    else:
        t = n // 10
        u = n % 10
        return tens[t] + ('' if u == 0 else ' ' + units[u])

even_numbers = [num for num in range(1, 51) if num % 2 == 0]
for n in even_numbers:
    print(f"{n}: {num_to_words(n)}")
```

```python
# group by department and count completed vs incomplete (no/other)
dept_counts = {}

with open('training.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    for i, row in enumerate(reader):
        if not row:
            continue
        # skip header if present
        if i == 0 and any(h.strip().lower() in ('name', 'department', 'status') for h in row):
            continue

        dept = row[1].strip() if len(row) > 1 and row[1].strip() else 'Unknown'
        status = row[2].strip().lower() if len(row) > 2 else ''

        completed = status in ('yes', 'y', 'completed', 'complete', 'true', '1')
        counts = dept_counts.setdefault(dept, {'completed': 0, 'incomplete': 0, 'total': 0})

        if completed:
            counts['completed'] += 1
        else:
            counts['incomplete'] += 1
        counts['total'] += 1

# print results sorted by department name
for dept in sorted(dept_counts):
    c = dept_counts[dept]
    print(f"{dept}: {c['completed']} completed, {c['incomplete']} incomplete (total {c['total']})")
```

```python
# load "training.csv" data in and modify the names to uppercase, then print each row to a new file "training_uppercase.csv"
import csv
with open('training.csv', mode='r', newline='', encoding='utf-8') as infile, \
     open('training_uppercase.csv', mode='w', newline='', encoding='utf-8') as outfile:
    csv_reader = csv.reader(infile)
    csv_writer = csv.writer(outfile)

    for row in csv_reader:
        if row:  # ensure the row is not empty
            row[0] = row[0].upper()  # convert the name (first column) to uppercase
        csv_writer.writerow(row)
```
