# Guided Worksheet 1: "Hello, Office!" 

Your first Copilot-Powered Program

Goal: Learn how to use GitHub Copilot in VS Code to create your first simple Python program that greets your team.

## Steps:

### Step 1: Open Your Workspace
1. Launch VS Code.
2.  Create a new folder on your desktop, e.g. **DxLabs_Week01**
3. Inside VS Code, open this folder (**File --> Open Folder**).
4. Create a new file named **hello_office.ipynb** (Python Notebook).

### Step 2: Check That Copilot Is Ready
1. Make sure you see the Copilot icon is on the lower right of VS Code.
2. If prompted, sign in to GitHub.
3. Ensure the Python kernel is active (you should see 'Python 3.x' at the top of your notebook).

### Step 3: Describe What You Want To Do
In the first code cell, type this comment:

`# a program that greets each person in the admin office by name`

Wait a second - Copilot will suggest code (gray text). Press TAB to accept.

### Step 4: Let Copilot Build the Program
Continue writing hints or comments that let Copilot fill in the code:

`# Loop through each name and print a greeting`

Press â–¶ Run Cell (or Shift+ENTER) to see the result.

### Step 5: Make it Yours
Experiment a little:
- Add more names to the list.
- Change the message text.
- Ask Copilot to add the current date:

`# Add today's date to the greeting`

- Ask Copilot to save greetings to a file:

`# Save greetings to greetings.txt`
